## [Computational Social Science] Project 5: Natural Language Processing
### Question 9 - Large Language Model and Text Generation

```python
from transformers import pipeline, set_seed
# Here, 'text-generation' tells the pipeline to create a text generation model, and model='gpt2'
# specifies that you want to use the GPT-2 model for this task.
generator = pipeline('text-generation', model='gpt2')
```

```python
set_seed(123)
input_text = "I am working on a project on natural language processing for a class on computational social science. This project"
output = generator(input_text, max_length=50, num_return_sequences=5)
for result in output:
    print(result['generated_text'])
    print("----")
```

```python
output = generator(input_text, max_length=50, num_return_sequences=5, temperature = 0.7)
for result in output:
    print(result['generated_text'])
    print("----")
```

```python
output = generator(input_text, max_length=100, num_return_sequences=5, temperature = 1.5)
for result in output:
    print(result['generated_text'])
    print("----")
```
