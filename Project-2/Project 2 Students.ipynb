#### d. Check for duplicate columns

There are a lot of columns in this data frame. Let's see if there are any are duplicates. Note that Pandas will not allow them to have the same exact column name, so they will likely be distinct on column name but will be copies otherwise.

```python
#
# Check for duplicate columns
#-----------
# I used Google to figure this out, and adapted this example for our purposes:  
# source: https://thispointer.com/how-to-find-drop-duplicate-columns-in-a-dataframe-python-pandas/ 
def getDuplicateColumns(df):
    '''
    Get a list of duplicate columns.
    It will iterate over all the columns in dataframe and find the columns whose contents are duplicate.
    :param df: Dataframe object
    :return: List of columns whose contents are duplicates.
    '''
    duplicateColumnNames = set()
    # Iterate over all the columns in dataframe
    for x in range(df.shape[1]):
        # Select column at xth index.
        col = df.iloc[:, x]
        # Iterate over all the columns in DataFrame from (x+1)th index till end
        for y in range(x + 1, df.shape[1]):
            # Select column at yth index.
            otherCol = df.iloc[:, y]
            # Check if two columns at x 7 y index are equal
            if col.equals(otherCol):
                duplicateColumnNames.add(df.columns.values[y])
    return list(duplicateColumnNames)

duplicateColumnNames = list(getDuplicateColumns(diabetes))
print('Duplicate Columns are as follows: ')
duplicateColumnNames
```

```python
#
# drop columns from duplicates list
#-----------
# now drop list of duplicate features from our df using the .drop() method
diabetes = diabetes.drop(columns=duplicateColumnNames) 
```

1. What is bias-variance tradeoff? Why is it relevant to machine learning problems like this one?

As the name implies, the bias-variance tradeoff denotes the relationship between model bias and model variance among various statistical models. Generally, as bias or variance decreases, the other increases, creating a trade-off. The term "bias" illustrates how well a statistical model "fits" or "models" the data. When bias is low, a statistical model fits the data very well, but when bias is high, a model may not appropriately represent its data. On the other hand, "variance" represents how well a statistical model represents new or resampled data from the same population. When variance is low, a statistical model may continue to appropriately represent new data, but when variance is high, new data may not be appropriately represented by the old model. In terms of the bias-variance tradeoff, as bias decrease (i.e., the model more closely represents the current data), variance increases (i.e., the model may not appropriately represent new data). In practice, we generally try to choose a statistical model that optimizes this this tradeoff with as low of bias and variance as possible.

2. Define overfitting, and why it matters for machine learning. How can we address it?

Overfitting occurs when a statistical model's bias is so low (i.e., the model very closely represents the data) that the model does not generalize well to new data (i.e., high variance). In many circumstances in machine learning, we try to develop models that demonstrate a relationship that generalizes to a population. In other words, if we resampled the same population, the findings from our statistical model wouldn't change by a significant degree. To avoid overfitting, we try to select a model that optimizes the relationship between bias and variances so that the model represents the data as well as possible while also generalizing to the population.
